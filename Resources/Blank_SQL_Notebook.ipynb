```python
%%sql

SELECT s.orderdate, s.quantity * s.netprice * s.exchangerate AS net_revenue
FROM sales s
LEFT JOIN customer c ON s.customerkey = c.customerkey
WHERE orderdate::date >= '2024-01-01'
limit 10;
```

```python
%%sql

SELECT orderdate, s.quantity * s.netprice * exchangerate AS net_revenue, c.givenname, c.surname, c.countryfull, c.continent
FROM sales s
LEFT JOIN customer c ON s.customerkey = c.customerkey
WHERE orderdate::date >= '2024-01-01'
limit 10;
```

```python
%%sql

SELECT orderdate, s.quantity * s.netprice * exchangerate AS net_revenue, c.givenname, c.surname, c.countryfull, c.continent
FROM sales s
LEFT JOIN customer c ON s.customerkey = c.customerkey
LEFT JOIN product p ON s.productKey = p.productKey
WHERE orderdate::date >= '2024-01-01'
limit 10;
```

```python
%%sql

SELECT orderdate, s.quantity * s.netprice * exchangerate AS net_revenue,
c.givenname,
c.surname,
c.countryfull,
c.continent,
p.productkey,
p.productname,
p.categoryname,
p.subcategoryname,
CASE WHEN s.quantity * s.netprice * exchangerate > 1000 THEN 'HIGH' ELSE 'LOW' END AS high_low
FROM sales s
LEFT JOIN customer c ON s.customerkey = c.customerkey
LEFT JOIN product p ON s.productKey = p.productKey
WHERE orderdate::date >= '2024-01-01'
limit 10;
```

```python
%%sql

SELECT orderdate, COUNT(customerKey) AS total_customers FROM sales GROUP BY orderdate ORDER BY orderdate
```

```python
%%sql

SELECT orderdate, COUNT(DISTINCT customerKey) AS total_customers FROM sales WHERE orderdate BETWEEN '2020-01-01' AND '2024-01-01'GROUP BY orderdate ORDER BY orderdate
```

```python
%%sql

SELECT orderdate, COUNT(DISTINCT s.customerKey) AS total_customer
FROM sales s
LEFT JOIN customer c ON c.customerKey = s.customerKey
WHERE orderdate BETWEEN '2020-01-01' AND '2024-01-01'
GROUP BY orderdate ORDER BY orderdate
```

```python
%%sql

SELECT orderdate,
COUNT(DISTINCT s.customerKey) AS total_customer,
COUNT(DISTINCT CASE WHEN c.continent = 'Europe' THEN s.customerKey END) AS eu_customer
FROM
sales s
LEFT JOIN customer c ON s.customerKey = c.customerKey
WHERE orderdate BETWEEN '2020-01-01' AND '2024-01-01'
GROUP BY orderdate
ORDER BY orderdate

```
